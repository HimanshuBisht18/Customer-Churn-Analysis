```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('Customer Churn.csv')


columns = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies',
       ]

n_cols = 3
batch_size = 6   # number of plots per figure

for batch in range(0, len(columns), batch_size):
    subcols = columns[batch: batch + batch_size]
    n_rows = (len(subcols) + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows*4))
    axes = axes.flatten()

    for i, col in enumerate(subcols):
        sns.countplot(x=col, data=df, ax=axes[i],hue="Churn")
        axes[i].set_title(f'Count Plot of {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Count')

    # remove unused axes or empty subplots
    for j in range(len(subcols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

#creates subplot we can concluded that Most customers who churned had no online security, no tech support, and no additional services like backup or protection. 
#Customers with fiber optics internet also show higher churn compared to DSL or no internet service.
